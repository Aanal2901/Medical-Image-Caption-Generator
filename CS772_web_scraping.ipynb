```python
import requests
from bs4 import BeautifulSoup
from nltk.corpus import gutenberg
```

```python
idx = gutenberg.fileids()
print(idx)
```

```python
for fileid in gutenberg.fileids():
  num_chars = len(gutenberg.raw(fileid))
  num_words = len(gutenberg.words(fileid))
  num_sents = len(gutenberg.sents(fileid))
  num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
  print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

```python
i=0
print(len(gutenberg.sents(idx[0])))
for sent in gutenberg.sents(idx[0]):
  i+=1
  full_sent = ''
  if sent[0]=="CHAPTER" or sent[0] == "VOLUME" or sent[0]=="[":
    continue

  for x in sent:
    if x=="." or x==",":
      full_sent=full_sent + x
    else:
      full_sent = full_sent + ' ' + x
  print(full_sent)
  if i==50:
    break
```

```python
emma = gutenberg.words(idx[0])
for i in range(0, len(emma), 10):
  word = []
  for j in range(10):
   word.append(emma[i+j])
  print(word)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

list(soup.children)

# find all occurrence of p in HTML
# includes HTML tags
# print(soup.find_all('p'))

# print('\n\n')

# return only text
# does not include HTML tags
for txt in soup.find_all('p'):
  print(txt.get_text())
```

```python
quote = []
quote_elements = soup.find_all('div', class_='quote')
```

```python
for quote_element in quote_elements:
  # extracting the text of the quote
  text = quote_element.find('span', class_='text').text
  print(text)
  # extracting the author of the quote
  author = quote_element.find('small', class_='author').text

  # extracting the tag <a> HTML elements related to the quote
  tag_elements = quote_element.find('div', class_='tags').find_all('a', class_='tag')

  # storing the list of tag strings in a list
  tags = []
  for tag_element in tag_elements:
      tags.append(tag_element.text)
  print(tags)
```
